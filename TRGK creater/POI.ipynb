```python
# 定义批量删除的查询
batch_size = 10000
while True:
    query = f"""
    MATCH (p:POI)
    WITH p LIMIT {batch_size}
    DETACH DELETE p
    RETURN COUNT(p) AS deleted_count
    """
    
    # 执行批量删除操作
    result = g.run(query).data()
    
    # 如果没有节点被删除，结束循环
    if result[0]['deleted_count'] == 0:
        break

    print(f"Deleted {result[0]['deleted_count']} nodes in this batch.")
```

```python
import geopandas as gpd
import os
```

```python
gdf_honeycomb = r"F:\paper2\result\dz_honeycomb_125.shp"
gdf_honeycomb = gpd.read_file(gdf_honeycomb)
print(gdf_honeycomb.crs)
```

```python
def get_file(folder_path):  #获取同一文件夹下所有的文件路径
    dir_file = os.listdir(folder_path)
    for path in dir_file:
        if path[-3:] == 'csv' or path[-4:] == 'xlsx' or path[-3:] == 'xls' or path[-3:] == 'shp':
            whole_path = (folder_path + "\\" + '{}').format(path)
            dir_xls.append(whole_path)
    return dir_xls
def get_name(folder_path):  #获取同一文件夹下所有的文件路径
    dir_file = os.listdir(folder_path)
    for path in dir_file:
        if path[-3:] == 'csv' or path[-4:] == 'xlsx' or path[-3:] == 'xls' or path[-3:] == 'shp':
            file_name_without_extension = os.path.splitext(path)[0]  # 获取文件名的主干，不包含扩展名
            folder_name.append(file_name_without_extension)
    return folder_name
```

```python
folder_path = r'E:\paper2\S-上海-SH\02-POI\23\Shp'
dir_xls = []
dir_xls = get_file(folder_path)
dir_xls
```

```python
BATCH_SIZE = 10000  # 定义批量导入的大小
# 输出文件保存路径
output_dir = r'F:\paper2\result\POI'
os.makedirs(output_dir, exist_ok=True)  # 确保输出文件夹存在
global_index = 0  # 定义全局计数器，从 0 开始


#创建POI实体及contain关系
poi_ls = []
relationship_ls = []

for folder in dir_xls:
    
    poi = gpd.read_file(folder)
    print(f"\n文件: {folder}")
    poi = poi.to_crs(gdf_honeycomb.crs)
    print(f"poi.crs:\n {poi.crs}")

    # 检查并重命名字段
    rename_mapping = {
        '经度': 'lng',  # 将 '经度' 字段重命名为 'lng'
        '纬度': 'lat'   # 将 '纬度' 字段重命名为 'lat'
    }
    poi.rename(columns=rename_mapping, inplace=True)
    
    # 使用批量化的空间连接操作将 `gdf_honeycomb` 的 `FID` 列加入到 `trajectory_matched` 中
    poi_matched = gpd.sjoin(poi, gdf_honeycomb[['geometry']], how='inner', predicate='within')
        
    if not poi_matched.empty:
        # 添加一个新的列 `global_index`，初始化为范围 [global_index, global_index + len(poi_matched)]
        poi_matched['global_idx'] = range(global_index, global_index + len(poi_matched))

        # 更新全局计数器
        global_index += len(poi_matched)

        print(poi_matched.head(1))
        
        for i in range(len(poi_matched)):
            # if i > 3:
            #     break
            poipt = poi_matched.iloc[i]
        
            node = Node('POI', name=int(poipt['global_idx']), type_main=poipt['行业大'], address=poipt['address'],
                        lng=float(poipt['lng']), lat=float(poipt['lat']), ad_name=poipt['adname'],
                        poi_name=poipt['name'], geometry=str(poipt['geometry']))
            poi_ls.append(node)

            # 获取匹配的多边形的 FID，直接从提前计算好的列中获取
            honeycomb_idx = poipt['index_right']
            # print(f"honeycomb_idx\n {honeycomb_idx}")
            honeycomb = honeycomb_nodes.get(honeycomb_idx)  # 从缓存中查找 `honeycomb` 节点
            # 构建 'contains' 关系
            contains_rel = Relationship(honeycomb, "contains", node)  # honeycomb 包含 POI
            relationship_ls.append(contains_rel)

            # 批量提交
            if len(poi_ls) >= BATCH_SIZE:
                subgraph = Subgraph(nodes=poi_ls, relationships=relationship_ls)
                tx = g.begin()
                tx.create(subgraph)
                g.commit(tx)
                print(f"已提交 {len(poi_ls)} 条 POI 节点及其关系")

                # 清空临时列表
                poi_ls.clear()
                relationship_ls.clear()

    # 提交剩余的记录
    if poi_ls:
        subgraph = Subgraph(nodes=poi_ls, relationships=relationship_ls)
        tx = g.begin()
        tx.create(subgraph)
        g.commit(tx)
        print(f"已提交剩余 {len(poi_ls)} 条 POI 节点及其关系")

        # 清空临时列表
        poi_ls.clear()
        relationship_ls.clear()

    

    # 去除输出文件名中的 "_WGS84"
    base_name = os.path.basename(folder).replace('_WGS84', '')
    output_file = os.path.join(output_dir, base_name)

    # 删除 `index_right` 列（如果存在）
    if 'index_right' in poi_matched.columns:
        poi_matched = poi_matched.drop(columns=['index_right'])
        print("已删除 index_right 列")

    # 保存结果
    poi_matched.to_file(output_file, driver='ESRI Shapefile', encoding='utf-8')
    print(f"已更新文件保存到: {output_file}")
```
