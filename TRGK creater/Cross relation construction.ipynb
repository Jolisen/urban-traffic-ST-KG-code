```python
from tqdm.notebook import tqdm

# 1. 初始化索引
road_index = road.sindex  # 建立R树索引

# 2. 逐个处理honeycomb节点并处理关系
for i in tqdm(range(len(honeybuffer)), total=len(honeybuffer), desc="Processing honeycomb-road relationships"):
    if i >= global_var:
        break

    polygon = honeybuffer.loc[i, 'geometry']

    # 通过R树进行简单包含查询
    possible_matches_index = list(road_index.intersection(polygon.bounds))
    possible_matches = road.iloc[possible_matches_index]

    # 精确查询，匹配在多边形内部的道路
    precise_matches_intersects = possible_matches[possible_matches.within(polygon)]

    k = precise_matches_intersects['index'].tolist()

    # 匹配知识图谱中的格网节点
    main_honeycomb = node_matcher.match('honeycomb', name=i).first()

    # 处理与当前 honeycomb 相交的道路节点
    for j in k:
        if j >= global_var:
            break

        sub_road = node_matcher.match('road', name=j).first()

        # 使用 Cypher 的 MERGE 语句保证关系的唯一性
        query = """
        MATCH (a:road {name: $sub}), (b:honeycomb {name: $main})
        MERGE (a)-[r:within]->(b)
        RETURN r
        """
        g.run(query, main=main_honeycomb['name'], sub=sub_road['name'])

# 3. 此时不再需要批量提交关系，Cypher 中的 MERGE 操作已经处理了唯一性问题
```

```python
from concurrent.futures import ThreadPoolExecutor
from tqdm.notebook import tqdm

# 1. 定义处理函数 cross_relation_construction
def cross_relation_construction(i):
    if i >= global_var:
        return []

    polygon = honeybuffer.loc[i, 'geometry']

    # 通过R树进行简单包含查询
    possible_matches_index = list(road_index.intersection(polygon.bounds))
    possible_matches = road.iloc[possible_matches_index]

    # 精确查询，匹配在多边形内部的道路
    precise_matches_intersects = possible_matches[possible_matches.within(polygon)]

    k = precise_matches_intersects['index'].tolist()

    # 匹配知识图谱中的格网节点
    main_honeycomb = node_matcher.match('honeycomb', name=i).first()

    relations = []
    for j in k:
        if j >= global_var:
            continue

        sub_road = node_matcher.match('road', name=j).first()

        # 使用Cypher查询检查两个节点之间是否存在"within"关系
        query = """
        MATCH (a:road {name: $sub})-[r:within]->(b:honeycomb {name: $main})
        RETURN count(r) > 0 as relation_exists
        """
        relation_exists = g.run(query, main=main_honeycomb['name'], sub=sub_road['name']).evaluate()

        # 只有在不存在关系时，才创建 within 关系
        if not relation_exists:
            relation = Relationship(sub_road, "within", main_honeycomb)
            relations.append(relation)

    return relations

# 2. 初始化索引
relation_ls = []
road_index = road.sindex  # 建立R树索引

# 3. 使用ThreadPoolExecutor进行并行处理，并加上进度条
with ThreadPoolExecutor(max_workers=120) as executor:  # 根据你的CPU核心数设置max_workers
    for relations in tqdm(executor.map(cross_relation_construction, range(len(honeybuffer))), total=len(honeybuffer), desc="Processing honeycomb-road relationships"):
        relation_ls.extend(relations)

# 4. 批量提交所有关系到数据库
if relation_ls:  # 确保有关系要提交
    subgraph = Subgraph(relationships=relation_ls)
    tx = g.begin()
    tx.create(subgraph)
    g.commit(tx)

# 5. 清空关系列表
relation_ls.clear()
```
